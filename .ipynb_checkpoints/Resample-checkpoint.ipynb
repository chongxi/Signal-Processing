```python
%pylab inline
import pandas as pd
from scipy.signal import resample
from IPython.html.widgets import *
```

```python
df = pd.read_csv('./S1Ach3.csv', header=None)
df.head(10)
```

```python
raw_data = df.values[:,df.values[0,:]==1.0]
```

```python
raw_data.shape
```

```python
fs = 1.0/320/1e-3; dt = 1.0/fs; T = len(raw_data)*dt
t = linspace(0,T,len(raw_data))
```

```python
@interact(i=(0,raw_data.shape[1]-1,1))
def view_trace(i=0):
    figure(figsize=(16,8))
    plot(t,raw_data[:,i])
    global x
    x = raw_data[:,i]
```

```python
data,t = resample(raw_data, len(x)*2, t, axis=0)
data.shape
```

```python
figure(figsize=(16,8))
plot(t,data[:,6])
fs = 1/(t[2]-t[1])
print fs
```
