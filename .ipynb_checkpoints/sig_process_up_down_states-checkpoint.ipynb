#Load raw_data from csv

```python
%pylab
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import pandas as pd
from IPython.html.widgets import *
from scipy.signal import filtfilt, butter
from scipy.signal import resample
```

```python
df = pd.read_csv('./S1Ach3.csv', header=None)
raw_data = df.values[:,df.values[0,:]==1.0]
print raw_data.shape
df.head(10)
```

```python
fs = 1.0/320/1e-3; dt = 1.0/fs; T = len(raw_data)*dt
t = linspace(0,T,len(raw_data))
data = raw_data
```

#Filtering and dynamically display the trace

```python
# Create an order 3 bandpass butterworth filter.
def butter_bandpass(lowcut, highcut, fs, order=3):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

@interact(i=(0,data.shape[1]-1,1))
def show_trace(i=0):
    fig = figure(figsize=(16, 6), dpi=100, facecolor='w', edgecolor='b')
    ax = fig.add_subplot(111)
    x = data[:,i]
    x = x-mean(x)
    b, a = butter_bandpass(0.01,0.3,fs)
    global idx
    idx = i
    global y
    y = filtfilt(b, a, x, padlen=150, padtype="even")  #pad method is very important in this case
    ax.plot(t,x,t,y,'r')
    return fig
```

#Amplitude Threshold

```python
def get_epoch(ud):
    from_0_to_1_idx = np.asarray(where(diff(ud)==1)) + 1
    from_1_to_0_idx = np.asarray(where(diff(ud)==-1))
    epoch = np.hstack((from_0_to_1_idx.T,from_1_to_0_idx.T))
    return epoch
    
@interact(thr=(min(y),max(y),0.02))
def get_up_down(t=fixed(t),y=fixed(y),thr=median(abs(y)/0.675)):
    # calculate epoch
    global epoch
    upstates = where(y>=thr)
    global ud
    ud = zeros(size(y))
    ud[upstates] = 1
    epoch = get_epoch(ud)
    # plot
    fig = figure(figsize=(16, 8), dpi=50, facecolor='w', edgecolor='b')
    ax1 = fig.add_subplot(211)
    ax1.plot(t,y)
    ax1.axhline(thr,ls='-.',c='m')
    ax1.plot(t[upstates],y[upstates],'r.')
    ax2 = fig.add_subplot(212)
    ax2.plot(t,ud,'g')
    ax2.plot(t[epoch],ud[epoch],'ro', markersize=4)
    ax2.set_ylim([min(ud)-0.1,max(ud)+0.1])
```

#Duration threshold

```python
def plot_upstates(t,y,epoch,duration_thr=0):
    fig = figure(figsize=(16, 8), dpi=50, facecolor='w', edgecolor='b')
    ax1  = fig.add_subplot(211)
    ax1.plot(t,y,'b',lw=2)
    ud = zeros(size(t))
    for epo in epoch:
        du = epo[1] - epo[0]
        if du <= duration_thr:
            ax1.plot(t[epo[0]:epo[1]],y[epo[0]:epo[1]],'go',lw=2)
        else:
            ax1.plot(t[epo[0]:epo[1]],y[epo[0]:epo[1]],'r-',lw=2)
            ud[epo[0]:epo[1]] = 1
    ax2  = fig.add_subplot(212)
    ax2.plot(t,y,'b',t,ud,'r')
    return fig
```

```python
@interact(duration_thr=(0,10,1))
def get_upstates(duration_thr=0, epoch=fixed(epoch)):
    global fig
    fig = plot_upstates(t,y,epoch,duration_thr)
    condition = epoch[:,1]-epoch[:,0] > duration_thr
    epoch = epoch[condition,:]
    global epoch_
    epoch_ = epoch
```

```python
fname = str(idx)+'.pdf'
fig.savefig(fname, bbox_inches='tight')
!open .
```
